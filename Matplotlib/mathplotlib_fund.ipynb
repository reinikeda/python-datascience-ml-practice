```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoint = np.array([0, 10])
ypoint = np.array([0, 100])

plt.plot(xpoint, ypoint)
plt.show()
```

```python
xpoints = np.array([0, 5, 7, 10])
ypoints = np.array([7, 84, 34, 77])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
plt.plot(ypoints)
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mfc = '#4CAF50', mec = 'r')
plt.show()
```

```python
plt.plot(ypoints, 'o--m')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dotted', color = 'g', linewidth = '5')
plt.show()
```

```python
plt.plot(ypoint)
plt.plot(ypoints)

plt.show()
```

```python
y = np.array([70, 45, 68, 59, 77, 50, 13, 24, 55])

plt.plot(y)

plt.title('Name of the plot')
plt.xlabel('the X label name')
plt.ylabel('the Y label name')

plt.grid(axis='y')

plt.show()
```
